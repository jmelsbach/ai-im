## Sigmoid Function

```python
import torch
from torch import sigmoid
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_function(func):
    x = torch.tensor(np.linspace(-8.,8., 100))
    y = torch.tensor([func(value) for value in x])
    plt.plot(x,y)
    plt.show()
```

```python
draw_function(sigmoid)
```

```python
a = torch.tensor(5.)
```

```python
sigmoid(a)
```

## Linear Functions

```python
def f1(x):
    return 2 * x + 3
```

```python
def f2(x):
    return 3 * x + 4
```

```python
def f1_f2(x):
    return f2(f1(x))
```

```python
def f3(x):
    return 6 * x + 13
```
